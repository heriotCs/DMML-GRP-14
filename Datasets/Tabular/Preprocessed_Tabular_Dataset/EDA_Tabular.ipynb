```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="muted", font_scale=1.1)

# Load Preprocessed Dataset
df = pd.read_csv("complete_preprocessed_plant_data.csv")

print("Dataset shape:", df.shape)
df.head()
```

```python
# Basic Info
print("\n--- Dataset Info ---")
print(df.info())

print("\n--- Missing Values ---")
print(df.isnull().sum())

print("\n--- Summary Statistics ---")
df.describe(include="all")
```

```python
# Distribution of target
plt.figure(figsize=(6,4))
sns.histplot(df['Health_Score'], kde=True, bins=20)
plt.title("Distribution of Health_Score")
plt.show()

# Relationship with a numeric feature
sns.scatterplot(x='Soil_Moisture_%', y='Health_Score', data=df)
plt.title("Soil Moisture vs Health Score")
plt.show()

# Relationship with a categorical feature
sns.boxplot(x='Fertilizer_Type_Chemical', y='Health_Score', data=df)
plt.title("Health Score by Fertilizer Type (Chemical)")
plt.show()
```

```python
print(df.columns)
```

```python
# Numerical Features
num_cols = df.select_dtypes(include=np.number).columns

# Histograms
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()


```

```python
# Individual histograms for each numeric feature
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

plt.figure(figsize=(10,8))
corr = df[num_cols].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    annot_kws={"size":8},   # smaller annotation text
    cbar_kws={"shrink": .8} # smaller colorbar
)

plt.title("Correlation Heatmap", fontsize=14)
plt.xticks(rotation=45, ha="right", fontsize=9)
plt.yticks(rotation=0, fontsize=9)
plt.tight_layout()
plt.show()

corr_target = corr["Health_Score"].sort_values(ascending=False)
print(corr_target)
```

```python
# 6. Categorical Features
cat_cols = [
    'Fertilizer_Type_Chemical',
    'Fertilizer_Type_Organic',
    'Fertilizer_Type_Liquid feed',
    'Fertilizer_Type_Compost',
    'Soil_Type_Loamy',
    'Soil_Type_Clay',
    'Sunlight_Exposure_Filtered sunlight through curtain',
    'Sunlight_Exposure_3h direct morning sun',
    'Sunlight_Exposure_Indirect light all day',
    'Pest_Presence_Aphids',
    'Pest_Presence_Fungus gnats',
    'Pest_Presence_Spider mites',
    'Pest_Severity_Low',
    'Pest_Severity_Moderate',
    'Pest_Severity_High',
    'Health_Notes_New bud'
]

for col in cat_cols:
    plt.figure(figsize=(7,4))
    sns.countplot(x=col, data=df)
    plt.title(f"Distribution of {col}")
    plt.xticks(ha="right") 
    plt.tight_layout()    
    plt.show()

    df[cat_cols].sum().sort_values(ascending=False)
```
