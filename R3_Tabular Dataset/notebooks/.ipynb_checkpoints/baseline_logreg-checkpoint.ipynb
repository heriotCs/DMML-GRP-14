## Baseline Model 1: Logistic Regression (Multi-Class Classification)

This notebook implements a baseline multinomial Logistic Regression model 
for predicting 'Health_Score' from the plant dataset.
This satisfies part of Coursework Requirement R3 — Baseline Training & Evaluation.


```python
# --- Libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
)

# --- Configuration ---
plt.style.use("seaborn-v0_8-darkgrid")
RANDOM_STATE = 42
```

```python
# --- Load training and testing data ---
train_path = "../data/training_preprocessed_plant_data.csv"
test_path  = "../data/testing_preprocessed_plant_data.csv"

train_df = pd.read_csv(train_path)
test_df  = pd.read_csv(test_path)

print("Training data shape:", train_df.shape)
print("Testing data shape:", test_df.shape)
```

```python
# --- Split features (X) and target (y) ---
X_train = train_df.drop("Health_Score", axis=1)
y_train = train_df["Health_Score"]

X_test = test_df.drop("Health_Score", axis=1)
y_test = test_df["Health_Score"]

print("X_train:", X_train.shape, " | y_train:", y_train.shape)
print("X_test :", X_test.shape, " | y_test :", y_test.shape)
```

```python
# --- Logistic Regression pipeline ---
log_reg = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(
        solver="saga",          # handles multi-class + collinearity better
        penalty="l2",           # ridge regularization
        C=0.05,                 # stronger regularization
        max_iter=3000,
        random_state=RANDOM_STATE,
        n_jobs=-1
    ))
])

# --- Train ---
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)

print("✅ Training complete.")
```

```python
# --- Metrics ---
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc:.4f}")

report = classification_report(y_test, y_pred, output_dict=True)
report_df = pd.DataFrame(report).transpose()
display(report_df.head())

# --- Save metrics ---
import os
os.makedirs("../results/metrics", exist_ok=True)
report_df.to_csv("../results/metrics/logreg_metrics.csv")
```

```python
# --- Confusion Matrix ---
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues", values_format="d")

plt.title("Logistic Regression – Confusion Matrix")
os.makedirs("../results/figures", exist_ok=True)
plt.savefig("../results/figures/logreg_confusion_matrix.png", bbox_inches="tight")
plt.show()
```

```python
# --- Compute RMSE for Logistic Regression ---
from sklearn.metrics import mean_squared_error
import numpy as np

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"✅ RMSE: {rmse:.4f}")

# --- Save classification metrics (with RMSE) ---
report_dict = classification_report(y_test, y_pred, output_dict=True)
report_df = pd.DataFrame(report_dict).transpose()

# Add RMSE as an extra row
report_df.loc["RMSE"] = {
    "precision": np.nan,
    "recall": np.nan,
    "f1-score": np.nan,
    "support": rmse
}

# Save metrics file
metrics_path = "../results/metrics/logreg_metrics.csv"
report_df.to_csv(metrics_path, index=True)
print(f"✅ Metrics (with RMSE) saved to: {metrics_path}")
```
